```python
import pandas as pd
```

```python
df = pd.read_csv('modelA.csv', header=0, index_col=0)
df
```

```python
df.columns
```

```python
df_percentage = pd.DataFrame()

df_percentage['Proto-Elamite'] = round((df['Proto-Elamite'] / df['Proto-Elamite'].sum()) * 100, 2)
df_percentage['Ur III'] = round((df['Ur III'] / df['Ur III'].sum()) * 100, 2)
df_percentage['Old Babylonian'] = round((df['Old Babylonian'] / df['Old Babylonian'].sum()) * 100, 2)
df_percentage
```

```python
df_percentage.transpose().plot()
```

```python
df_percentage['Proto-Elamite'].plot.pie()
```

```python
df_percentage['Ur III'].plot.pie()
```
